```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel("http://share.tn.gov/sos/election/results/2010-11/November2010.xls")
df.head()
```

```python
df.columns
```

```python
df = df.rename(columns={
    "District":"district",
    "COUNTY":"county",
    "OFFICENAME":"office",
    "Precinct Sequence":"precinct"
  }
)
df["county"] = df.county.apply(lambda s: s.capitalize())
df.head()
```

```python
df.office.unique()
```

```python
df = df.loc[df.office == "Governor", :]
df["district"] = df.district.fillna("")
df = df.sample(20)
df.head()
```

## Flatmap rows

```python
def flatten(df_x):
  d = {'candidate_party': [], 'votes': []}
  tups = [("BName{0}".format(i), "Tally{0}".format(i)) for i in range(1,11)]
  for name, tally in tups:
      d['candidate_party'].extend(df_x[name])
      d['votes'].extend(df_x[tally])
  return pd.DataFrame(d)
```

```python
df_new = df.groupby(["county", "precinct", "office", "district", "Candidate Group"]).apply(flatten).reset_index()
df_new = df_new.dropna(subset=["candidate_party"])
df_new
```

## Splits party from candidate name

```python
import re

class Splitter():
  def __init__(self):
    self.cand_party = re.compile(r"^\.\s(.*)\s-\s([A-Z])$")
    self.writein_cand = re.compile(r"^\.\s(Write-in)\s-\s(.*)")

  def get_candidate_party(self, s):
    if not isinstance(s, str):
      raise ValueError("Unrecognized input format")
    m = self.cand_party.match(s)
    if m is not None:
      name, party = m.groups()
      return (name, party)
    else:
      m = self.writein_cand.match(s)
      if m is not None:
        party, name = m.groups()
      else:
        raise ValueError("Unrecognized input format")
        
    return (name, party)

splitter = Splitter()
```

```python
df_new["candidate"], df_new["party"] = zip(*df_new.candidate_party.map(splitter.get_candidate_party))

df_new
```

```python
df_new.groupby(["county", "candidate", "party", "office", "district"]).agg({
    "votes": np.sum,
  }).reset_index().sort_values(["candidate", "county"])
```
